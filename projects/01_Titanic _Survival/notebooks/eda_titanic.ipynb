```python
#--- Libraries ---#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Style Configuration (with a reset) ---

# 1. Reset Matplotlib to its default style.
# This is the key step to override any dark theme from your IDE.
plt.style.use('default')

# 2. Now, apply the Seaborn theme you want.
# This will now work as expected on top of the default settings.
sns.set_theme(style="whitegrid", context="talk", palette="pastel")

# After running this cell, all subsequent plots in your notebook
# should now correctly have the light theme.
```

```python
#--- Style Configuration ---#
sns.set_style("whitegrid")
plt.style.use('seaborn-v0_8-deep')
sns.set_palette("pastel")
sns.set_context("paper")
```

```python
#--- Data charge ---#
file_path = '../datasets/train.csv'
df=pd.read_csv(file_path)
```

```python
#--- Check the first 5 rows ---#
df.head()
```

```python
#--- Technical Summary ---#
df.info()
```

```python
#--- Statistical Summary ---#
df.describe()
```

```python
# --- Create a clean and clear plot (Simplified Version) --- #

# 1. Set a professional theme for the plot.
sns.set_theme(style="whitegrid", context="talk")

# 2. Set the size of the figure.
plt.figure(figsize=(10, 7))

# 3. Create the main plot. This is the most important command.
ax = sns.countplot(
    x='Survived',
    data=df,
    palette='pastel'
)

# 4. Add clear titles and labels to explain what the plot is about.
ax.set_title('Survival on the Titanic', fontsize=18)
ax.set_xlabel('Outcome', fontsize=14)
ax.set_ylabel('Number of Passengers', fontsize=14)

# 5. Make the labels on the x-axis (the "ticks") more descriptive.
ax.set_xticklabels(['Did Not Survive', 'Survived'])

# 6. Display the final plot.
plt.show()
```

```python
#--- A plot to see how many people from each class survived ---#
plt.figure(figsize=(10,6))

ax = sns.countplot(
    x='Pclass',
    data=df,
    hue='Survived',
    palette='viridis',
)

plt.title('Survival Count by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.legend(
    title='Status',
    labels=['Not Survived', 'Survived'],
)

ax.set_xticklabels(['1st Class', '2nd Class', '3rd Class'])

plt.show()
```

```python
#--- Now , let's go with the survival rates for men and women ---#
plt.figure(figsize=(10,6))
sns.countplot(
    x='Sex',
    hue='Survived',
    palette='plasma',
    data=df,
)

plt.title('Survival Count by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.legend(
    title='Status',
    labels=['Not Survived', 'Survived'],
)
plt.show()
```

```python
#--- Grid of plots: one for passengers who survived, one for those who did not ---#

g = sns.FacetGrid(df, col='Survived', height=6)

g.map(sns.histplot, 'Age', bins=30, kde=True)

g.set_axis_labels("Age", "Number of Passengers")
g.axes[0,0].set_title('Did not Survive')
g.axes[0,1].set_title('Survived')

plt.show()
```

```python
#--- Now we are going to check the relation with the ticket prize ---#
plt.figure(figsize=(10,6))

sns.boxplot(x='Survived', y='Fare', data=df, palette='viridis')

#--- Add titles and labels ---#

plt.title('Fare Distribution by Survival Status', fontsize=14)
plt.xlabel('Survival Status', fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.xticks([0,1],['Not Survived', 'Survived'])
plt.ylim(0,300)
plt.show()
```
